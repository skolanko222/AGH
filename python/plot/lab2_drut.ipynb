```python
import matplotlib.pyplot as plt
import numpy as np


y = [0.52,1.2875,1.9375,2.437,2.91]
x = [10,20,30,40,50]

coef  = np.polyfit(x,y,3)
poly1d_fn = np.poly1d(coef) 

plt.plot(x, y, 'yo',poly1d_fn( [i for i in range(0,51)])) 
plt.xlabel('t [\u03BC s]')
plt.ylabel('A [V]')
plt.grid()

plt.show()
```
