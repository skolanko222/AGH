```python
import matplotlib.pyplot as plt
import numpy as np
#n  5
arr = []
with open("data_lab7.txt") as file:
	for line in file.readlines():
		arr.append([float(i) for i in line.split(' ') if i != '\n' and i != ''])

print(arr[0])
plt.plot(arr[0], arr[1],markersize=0.5) 
plt.plot(arr[2], arr[3],markersize=0.5) 
plt.plot(arr[4], arr[5],markersize=0.5)
plt.gca().legend(('exp(-x^2) n=5',))
plt.ylabel('x')
plt.xlabel('y')
plt.grid()


#plt.show()

```
