```python
import matplotlib.pyplot as plt
import numpy as np
def to_float(x):
	try:
		return float(x)
	except ValueError:
		return 0
arr = []
with open("data_lab10.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i is not '\n'])
# U1
#make dots smaller
plt.rcParams['lines.markersize'] = 2
plt.plot(arr[0], arr[1], 'yo') 
plt.ylabel('f(x)')
plt.xlabel('x')
plt.grid()
plt.savefig('u1a.png')

#plt.show()

```
