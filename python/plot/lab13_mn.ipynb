```python
import matplotlib.pyplot as plt
import numpy as np
def to_float(x):
	try:
		return float(x)
	except ValueError:
		return 0
arr = []
with open("data_lab13.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i is not '\n'])

print(arr)
plt.plot(arr[0], arr[1],markersize=0.5, color='red', label='krlik')
plt.plot(arr[0], arr[2],markersize=0.5, color='blue', label='zombie')
plt.legend(loc="upper left")
plt.ylim(0.5, 2.)
plt.xlim(0, 40)

plt.grid()
```
