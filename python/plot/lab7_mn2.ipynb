```python
import matplotlib.pyplot as plt
import numpy as np
#n  5
arr = []
with open("data_lab7_pochodne.txt") as file:
	for line in file.readlines():
		arr.append([float(i) for i in line.split(' ') if i != '\n' and i != ''])

plt.plot(arr[0], arr[1],markersize=0.5, label='2 pochodna ze wzoru')
plt.plot(arr[2], arr[3],markersize=0.5, label='2 pochodna z ilorazu różnicowego')
plt.legend(loc="upper left")
plt.ylim(-1, 1.5)

plt.ylabel('x')
plt.xlabel('y')
plt.grid()


#plt.show()

```
