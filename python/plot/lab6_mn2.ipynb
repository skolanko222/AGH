```python
import matplotlib.pyplot as plt
import numpy as np
#n  5
arr = []
with open("data_lab6_a2.txt") as file:
	for line in file.readlines():
		arr.append([float(i) for i in line.split(' ') if i is not '\n'])

#print(arr[0])
plt.plot(arr[0], arr[1],markersize=0.5) 
plt.gca().legend(('n=5',))
plt.ylabel('x')
plt.xlabel('y')
plt.grid()


#plt.show()

```
