```python
import matplotlib.pyplot as plt
import numpy as np


x = [0.0011,1.0166,2.0324,3.0504,4.0678,5.0874,6.11]
y = [0.0002,0.9883,1.997,2.995,3.993,4.992,5.99]

coef  = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 

plt.plot(x, y, 'yo',poly1d_fn(x)) 
plt.ylabel('U [V]')
plt.xlabel('I [mA]')
plt.grid()

plt.show()
```
