```python
import matplotlib.pyplot as plt
import numpy as np

arr = []
with open("data_lab5.txt") as file:
	for line in file.readlines():
		arr.append([i for i in line.split(' ') if i is not '\n'])

print(arr)
fig, ax = plt.subplots()
fig.canvas.draw()
ax.set_xticklabels(range(1,11))
for i in arr:
	plt.plot([_ for _ in range(len(i))], [float(_) for _ in i],markersize=0.5) 
plt.gca().legend(('D(c)','B(C)'))
plt.ylabel('wartości')
plt.xlabel('krok C')
plt.grid()


#plt.show()

```
