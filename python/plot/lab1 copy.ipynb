```python
import matplotlib.pyplot as plt
import numpy as np


x = [0.454,0.412,0.370,0.330,0.286]
y = [1.343,1.275,1.199,1.135,1.075]

coef  = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 

plt.plot(x, y, 'yo') 
plt.ylabel('T [s]')
plt.xlabel('l [m]')
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error

x = [0.454,0.412,0.370,0.330,0.286]
y = [1.343,1.275,1.199,1.135,1.075]
y2= [i**2 for i in y]
print("wartości zmierzone: ",y2)

coef, cov = np.polyfit(x,y2,1,cov=True)
poly1d_fn = np.poly1d(coef)
y_fitted = [poly1d_fn(i) for i in x]
print("wartości funkcji przybliżonej: ",y_fitted)

u_a = mean_squared_error(y,y_fitted)
print("u_a: ", u_a)

plt.plot(x, y2, 'yo', x, poly1d_fn(x), '--k',markersize=10) 
plt.ylabel('T^2 [s]')
plt.xlabel('l [m]')
plt.grid()


plt.show()

```
