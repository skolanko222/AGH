```python
import matplotlib.pyplot as plt
import numpy as np

arr = []
with open("data_lab4.txt") as file:
	for line in file.readlines():
		arr.append([i for i in line.split(' ') if i is not '\n'])

print(arr)

for i in arr:
	plt.plot([_ for _ in range(len(i))], [float(_) for _ in i],markersize=0.5) 

plt.ylabel('Wartości lambdy')
plt.xlabel('liczba iteracji')
plt.grid()


plt.show()

```
