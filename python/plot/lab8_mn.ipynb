```python
import matplotlib.pyplot as plt
import numpy as np
#n  11
arr = []
with open("data_lab8.txt") as file:
	for line in file.readlines():
		arr.append([float(i) for i in line.split(' ') if i != '\n' and i != ''])

plt.plot(arr[0], arr[1])

plt.ylabel('y')
plt.xlabel('x')
plt.grid()
```
