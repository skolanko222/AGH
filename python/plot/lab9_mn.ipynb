```python
import matplotlib.pyplot as plt
import numpy as np
#n  11
arr = []
rel_val = []
with open("data_lab9_r.txt") as file:
	for line in file.readlines():
		rel_val.append([float(i) for i in line.split(' ') if i != '\n' and i != ''])
with open("data_lab9.txt") as file:
	for line in file.readlines():
		arr.append([float(i) for i in line.split(' ') if i != '\n' and i != ''])

plt.plot(arr[0], arr[1],markersize=0.5, label='aprox. n=5')
plt.plot(rel_val[0], rel_val[1],markersize=0.5, label='y=f(x)')
plt.legend(loc="upper left")

print(len(arr[2]))
print(len(arr[3]))
plt.ylabel('y')
plt.xlabel('x')

plt.grid()
```
