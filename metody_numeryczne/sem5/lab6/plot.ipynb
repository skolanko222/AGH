```python
import matplotlib.pyplot as plt
import numpy as np
def to_float(x):
	try:
		return float(x)
	except ValueError:
		return 0
arr = []
with open("plots/matrix5a.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i != '\n'])
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
plt.imshow(arr, cmap='seismic',vmin=-10,vmax=10)
plt.colorbar()
```
