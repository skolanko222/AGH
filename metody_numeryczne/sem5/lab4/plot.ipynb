```python
import matplotlib.pyplot as plt
import numpy as np
def to_float(x):
	try:
		return float(x)
	except ValueError:
		return 0
arr = []
with open("plots/globalErrorPlot.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i is not '\n'])

plt.plot(arr[0][1:], arr[1][1:],markersize=0.5, color='blue', label='omega = 1')
plt.plot(arr[2][1:], arr[3][1:],markersize=0.5, color='orange', label='omega = 0.6')
plt.legend()
plt.xscale('log')
```

```python
arr = []
with open("plots/localErrorPlot.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i is not '\n'])

plt.plot(arr[0][1:], arr[1][1:],markersize=0.5, color='orange', label='omega = 1.0')
plt.plot(arr[2][1:], arr[3][1:],markersize=0.5, color='blue', label='omega = 1.4')
plt.plot(arr[4][1:], arr[5][1:],markersize=0.5, color='yellow', label='omega = 1.8')
plt.plot(arr[6][1:], arr[7][1:],markersize=0.5, color='green', label='omega = 1.9')

plt.legend()
plt.grid()
plt.xscale('log')
```

```python
arr = []
with open("plots/globalErrorMatrix.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i != '\n'])
plt.figure()
plt.xlabel('x')
plt.ylabel('y')
plt.imshow(arr, cmap='jet',vmin=0)

plt.colorbar()
```
