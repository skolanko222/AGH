```python
import matplotlib.pyplot as plt
import numpy as np
def to_float(x):
	try:
		return float(x)
	except ValueError:
		return 0
arr = []
with open("plots/globalErrorPlot.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i is not '\n'])

plt.plot(arr[0][1:], arr[1][1:],markersize=0.5, color='blue', label='omega = 1')
plt.plot(arr[2][1:], arr[3][1:],markersize=0.5, color='orange', label='omega = 0.6')
plt.legend()
plt.xlabel('iteracja')
plt.ylabel('S')
plt.title('Relaksacja globalna')
plt.xscale('log')
```

```python
arr = []
with open("plots/localErrorPlot.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i is not '\n'])

plt.plot(arr[0][1:], arr[1][1:],markersize=0.5, color='orange', label='omega = 1.0')
plt.plot(arr[2][1:], arr[3][1:],markersize=0.5, color='blue', label='omega = 1.4')
plt.plot(arr[4][1:], arr[5][1:],markersize=0.5, color='yellow', label='omega = 1.8')
plt.plot(arr[6][1:], arr[7][1:],markersize=0.5, color='green', label='omega = 1.9')
plt.xlabel('iteracja')
plt.ylabel('S')
plt.title('Relaksacja lokalna')
plt.legend()
plt.grid()
plt.xscale('log')
```

```python
arr = []
with open("plots/globalErrorMatrix1.txt") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i != '\n'])
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
plt.imshow(arr, cmap='jet',vmin=0)
plt.colorbar()
fig.canvas.draw()
labels = [to_float(item.get_text())/10 for item in ax.get_xticklabels()]
ax.set_xticklabels(labels)
labels = [to_float(item.get_text())/10 for item in ax.get_yticklabels()]
ax.set_yticklabels(labels)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relaksacja globalna w=1')
plt.gca().invert_yaxis()
```
