```python
import matplotlib.pyplot as plt
import numpy as np
def to_float(x):
	try:
		return float(x)
	except ValueError:
		return 0.
arr = []
with open("data/data_100.dat") as file:
	for line in file.readlines():
			arr.append([to_float(i) for i in line.split(' ') if i != '\n'])

arr = np.reshape(arr, (41, 41))
# resize array to 40x40
arr = arr[0:40, 0:40]
# flip x and y axis
arr = np.flip(arr, 0)
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
plt.imshow(arr, cmap='seismic')
plt.colorbar()
```
