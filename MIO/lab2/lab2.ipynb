```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Proszę przeprowadzić generację danych uczących (xu, ku) oraz testujących (xt, kt) za pomocą
# generatora liczb pseudolosowych o rozkładach N([0, 0], 1) dla pierwszej klasy oraz N([2, 0], 1)
# dla drugiej o liczność każdej z prób wzorców 50 elementów. Próba testowa niech zawiera po
# 1000 elementów w każdej z klas.


def generate_data(mean, std, n):
	return np.array([[random.gauss(mean[0], std), random.gauss(mean[1], std)] for _ in range(n)])
# neuron - funkcja aktywacji f

def fun(net):
	return 1 if net > 0 else 0
	

def main():
	# dane uczące
	xu1 = generate_data([0, 0], 1, 50)
	xu2 = generate_data([5, 0], 1, 50)
	ku1 = np.array([0 for _ in range(50)])
	ku2 = np.array([1 for _ in range(50)])
	# dane testujące
	xt1 = generate_data([0, 0], 1, 1000)
	xt2 = generate_data([5, 0], 1, 1000)
	kt1 = np.array([0 for _ in range(1000)])
	kt2 = np.array([1 for _ in range(1000)])

	# wykres danych uczących
	plt.scatter(xu1[:, 0], xu1[:, 1], c='r', marker='x', label='klasa 1')
	plt.scatter(xu2[:, 0], xu2[:, 1], c='b', marker='o', label='klasa 2')
	plt.legend()
	plt.show()

	# wykres danych testujących
	plt.scatter(xt1[:, 0], xt1[:, 1], c='r', marker='x', label='klasa 1')
	plt.scatter(xt2[:, 0], xt2[:, 1], c='b', marker='o', label='klasa 2')
	plt.legend()
	plt.show()

if __name__ == '__main__':
	main()
```
